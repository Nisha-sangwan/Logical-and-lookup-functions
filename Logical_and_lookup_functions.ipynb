applicable.

 **solution:**

 To determine the commission rate for 45,000 in sales:

 step 1:  Use the Formula: `**=VLOOKUP(45000, A2:B5, 2, TRUE)**`.

 step 2: Since 45,000 falls between 20,000 and 50,000, the "True" argument returns the rate for the next lowest value.

 step 3: **Result**: 10%.
